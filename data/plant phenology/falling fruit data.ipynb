```python
import csv
from datetime import datetime
import pandas as pd 
```

# malus data

```python
def convert_date(date_string):
    date_object = datetime.strptime(date_string, "%m/%d/%Y")
    day_of_year = date_object.timetuple().tm_yday
    year = date_object.year
    return day_of_year, year

input_file = "falling fruit malus data.csv"
output_file = "ff malus.csv"

with open(input_file, "r") as file:
    reader = csv.reader(file)
    header = next(reader)  # Read the header row
    header.extend(["Day of Year", "Year"])
    
    # Store modified rows for preview
    modified_rows = []
    
    for row in reader:
        date_string = row[3]
        day_of_year, year = convert_date(date_string)
        row.extend([day_of_year, year])
        modified_rows.append(row)
        
# Print header
print("Header:")
print(header)

# Print preview of modified rows
print("\nPreview of modified rows:")
for row in modified_rows:
    print(row)

# Write modified rows to output CSV file
with open(output_file, "w", newline="") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(header)
    writer.writerows(modified_rows)

print("\nCSV file exported successfully.")
```

```python
# merging ff and npn/pep data

npn_pep = pd.read_csv("/Users/claudiado/Documents/falling fruit/malus/malus_final.csv")
ff = pd.read_csv("/Users/claudiado/Documents/falling fruit/falling fruit data/ff malus.csv")

malus = pd.concat([npn_pep, ff], axis = 0)

print(malus.head())
print(malus)

malus.to_csv('/Users/claudiado/Documents/falling fruit/malus.csv', index = False)
```
